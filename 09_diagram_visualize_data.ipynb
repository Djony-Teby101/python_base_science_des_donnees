# TRACER DES GRAPHIQUES

```python
import pandas as pd
df=pd.read_csv("data.csv")
```

```python
df.price_paid=df.price_paid.apply(lambda x: x.replace("$",""))
df.price_paid=df.price_paid.astype(float)
```

```python
df.price_paid.dtype
```

```python
df.groupby("date")["price_paid"].sum().plot(figsize=[20,10])
```

# AUTRE type de graphique.

```python
df.groupby("gender")["price_paid"].mean().plot.pie(legend=True)
```

```python
df.groupby("gender")["price_paid"].mean()
```

```python
df.groupby("country")["price_paid"].sum().plot.bar(rot=0,legend=True)
```
